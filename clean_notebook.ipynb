```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Invoices-with-Merged-Categories-and-Merged-Amounts.xlsx')
```

```python
df.head()
```

```python
df[['Category','Amount','Date']] = df[['Category','Amount','Date']]\
                                    .apply(lambda x: x.str.split('|'))
```

```python
df.head()
```

```python
def unnesting(df, explode):
    idx = df.index.repeat(df[explode[0]].str.len())
    df1 = pd.concat([
        pd.DataFrame({x: np.concatenate(df[x].values)}) for x in explode], axis=1)
    df1.index = idx

    return df1.join(df.drop(explode, 1), how='left')


new_df = unnesting(df,['Category', 'Amount', 'Date'])
```

```python
new_df = new_df[['Order ID', 'Category','Amount','Date']]
```

```python
new_df.reset_index(inplace=True, drop=True)
```

```python
new_df.head()
```

```python
new_df.Date = pd.to_datetime(new_df['Date'])
```

```python
new_df.to_excel('cleaned_data.xlsx')
```
